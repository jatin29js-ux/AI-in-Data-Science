Data Visulization with Matplotlib


```python
#import libraries
import matplotlib.pyplot as plt
```

```python
x=["Raghav","Jatin","Ayush","Karan","Ankit"]
y=[88,56,98,56,78]
plt.hist(y)
plt.figure(figsize=(5,5))

plt.show()
plt.plot(x,y,marker="o",linestyle=":",linewidth=2,color="y")

plt.xlabel("Name of subject")
plt.ylabel("Marks")
plt.title("Marks of Students")
plt.grid("True")
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
dict={
    'marks':[88,56,98,56,78]
}
df=pd.DataFrame(dict)
df.index=["Raghav","Jatin","Ayush","Karan","Ankit"]

plt.plot(df["marks"],marker="o",linestyle=":",linewidth=2,color="y")
plt.xlabel("Name of subject",color="g",fontsize=16)
plt.ylabel("Marks",color="r",fontsize=15)
plt.title("Marks of Students")
plt.grid("True")

plt.show()
```

HISTOGRAM

```python
plt.hist(y,bins=10,color="r",edgecolor="b",alpha=1)
plt.xlabel("Name of subject")
plt.ylabel("Marks")
plt.title("Marks of Students")
plt.grid(axis='x')
plt.show()
```

pichart

```python
sub=["physics","chemistry","maths","english","hindi"]
marks=[20,21,19,18,22]
plt.pie(marks,labels=sub,autopct="%1.1f%%",startangle=90,explode=(0,0,0,0,0.2))
plt.show()
```

```python
import matplotlib.pyplot as plt
x=["Raghav","Jatin","Ayush","Karan","Ankit"]
y=[88,56,98,56,78]
plt.plot(x,y,marker="o",linestyle=":",linewidth=2,color="y")

plt.xlabel("Name of Students")
plt.ylabel("Marks")
plt.title("Name of Students vs Marks")
for i in range(len(x)):
    plt.text(i + 0.1, y[i] + 0.1, x[i], fontsize=12,color="g")
plt.show()
```

Q.Can we draw two lines in a single graph?

Yes, we can draw

```python
z = [87,32,14,76,45]
plt.plot(x,y,color ="g", marker="o")
plt.plot(x,z,color="r",marker="o")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.title("Students vs Marks")
plt.scatter(x, y, color="b")
plt.show()
```

```python
#Add legend in your graph
z = [88,66,55,54,76]
plt.plot(x,y,color ="r", marker="o",label="English")
plt.plot(x,z,color="b",marker="o",label="SST")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.title("Students vs Marks")
plt.scatter(x, y, color="b")
plt.legend()
plt.show()
```
