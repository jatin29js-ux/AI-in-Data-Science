Data Visualization with Plotly


```python
#Import libraries
import pandas as pd
import plotly.express as px
import matplotlib.pyplot as plt
```

```python
India_data=px.data.gapminder().query("country=='India'")
India_data.head()
```

```python
#Bar_graph
graph=px.bar(India_data,x='year',y='pop',title='Population of India',labels={'pop':'Population','year':'Year'},color='pop',hover_data=['lifeExp','gdpPercap'],text='pop')
graph.update_traces(texttemplate='%{text:.2s}',textposition='inside')
graph.update_layout(uniformtext_minsize=8,uniformtext_mode='hide',xaxis_tickangle= -45,yaxis_type='log',template='plotly_dark')

graph.show()

```

```python
#India population over the inida
graph=px.line(India_data,x='year',y='pop',title='Population of India',labels={'pop':'Population','year':'Year'},color='pop',hover_data=['lifeExp','gdpPercap'],text='pop')
graph.show()
```

```python
#Scatter plot
graph=px.scatter(India_data,x='gdpPercap',y='lifeExp',size='pop',title='Population of India',labels={'pop':'Population','year':'Year'},color='pop',hover_data=['lifeExp','gdpPercap'],text='pop')
graph.update_layout(template='plotly_dark')

graph.update_traces(texttemplate='%{text:.2s}')
graph.show()
```

```python
#Piechart
px.pie(India_data,values='pop',names='year',title='Population of India',hole=0.1)
```

```python
#Histogram
px.histogram(India_data,x='pop')
```

```python
#Box plot
graph=px.box(India_data,y='pop',points="all",color_discrete_sequence=['red'])
graph.update_traces(boxmean=False,jitter=0.1,marker=dict(size=5),line=dict(width=1),opacity=0.10)
graph.show()
```

```python
#Violin_plot
px.violin(India_data,y='pop')
```

```python
#Heatmap
px.density_heatmap(India_data,y='pop')
```

```python
px.scatter_matrix(India_data,dimensions=['pop','lifeExp','gdpPercap'])
```

```python
#Regressionplot
px.scatter(India_data,y='pop',trendline='ols')
```
